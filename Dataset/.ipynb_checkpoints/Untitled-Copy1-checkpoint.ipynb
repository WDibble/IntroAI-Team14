```python
disaster_types = ['Drought', 'Earthquake', 'Epidemic', 'Extreme temperature', 'Flood', 'Storm', 'Volcanic activity', 'Wildfire']
```

```python
df2 = df[df['Country'].isin(country_list)]
```

```python
df2 = df2[df2['Disaster Type'].isin(disaster_types)]
```

```python
df2 = df2.drop('Seq', axis=1)
df2 = df2.drop('Glide', axis=1)
df2 = df2.drop('Disaster Group', axis=1)
df2 = df2.drop('Disaster Subsubtype', axis=1)
df2 = df2.drop('Event Name', axis=1)
df2 = df2.drop('OFDA Response', axis=1)
df2 = df2.drop('Appeal', axis=1)
df2 = df2.drop('Associated Dis', axis=1)
df2 = df2.drop('Associated Dis2', axis=1)
df2 = df2.drop('Declaration', axis=1)
df2 = df2.drop('Aid Contribution', axis=1)
df2 = df2.drop('Local Time', axis=1)
df2 = df2.drop('Adm Level', axis=1)
df2 = df2.drop('Admin1 Code', axis=1)
df2 = df2.drop('Admin2 Code', axis=1)
df2 = df2.drop('Geo Locations', axis=1)
```

```python
df2 = df2.dropna(subset=["Start Day"])
df2 = df2.dropna(subset=["Start Month"])
```

```python
df2["CPI"].fillna("100", inplace = True)
```

```python
df2["Total Damages Adjusted ('000 US$)"] = (df2["Total Damages ('000 US$)"].astype(float) / (df2["CPI"].astype(float)/100)).round(2)
```
